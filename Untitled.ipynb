## 2- Importing and Understanding the Data

```python
# Load dataset
data = pd.read_csv('Walmart_Store_sales.csv')
data.head()
```

This data set cointains 8 different feature and 6435 instances
The different features are : 
Store: The number of the store
Date: The date of the week
Weekly_Sales: the sales in week for a specific store
Holiday_Flag: a binary value to indicate if that week has a holiday
Temperature: The temperature for the week of sales
Fuel_Price: cost of fuel in the region of the store
CPI:  Prevailing consumer price index
Unemployment: Prevailing unemployment rate

By checking the $Walmart Store sales.csv$ file the date column had different ways of presenting the different dates. We will start by standerizing it.

```python
data['Date'] =  pd.to_datetime(data['Date'])
```

```python
# creating day, month, and year columns
data["Day"]= pd.DatetimeIndex(data['Date']).day
data['Month'] = pd.DatetimeIndex(data['Date']).month
data['Year'] = pd.DatetimeIndex(data['Date']).year
```

```python
data.head()
```

In this data set 4 types of holidays exists:

- The super bowl(american football)
- Labour day
- Thanks giving
- Christmas

```python
Super_Bowl   = ['2010-12-02','2011-11-02','2012-10-02']   
Labor_Day    = ['2010-10-09','2011-09-09','2012-07-09']
Thanksgiving = ['2010-11-26','2011-11-25','2012-11-23'] 
Christmas    = ['2010-12-31','2011-12-30','2012-12-28'] 

data['Super_Bowl'] = data['Date'].dt.date.astype(str).apply([lambda x: 1 if x in Super_Bowl else 0])
data['Labor_Day'] = data['Date'].dt.date.astype(str).apply([lambda x: 1 if x in Labor_Day else 0])
data['Thanksgiving'] = data['Date'].dt.date.astype(str).apply([lambda x: 1 if x in Thanksgiving else 0])
data['Christmas'] = data['Date'].dt.date.astype(str).apply([lambda x: 1 if x in Christmas else 0])
```

```python
data['Date'].dt.date.astype(str)[0]
```

```python
data.head()
```

```python
fig, axs = plt.subplots(ncols=4,figsize=(20,10))

sns.barplot(x='Super_Bowl', y='Weekly_Sales', data=data, ax=axs[0])
sns.barplot(x='Labor_Day', y='Weekly_Sales', data=data, ax=axs[1])
sns.barplot(x='Thanksgiving', y='Weekly_Sales', data=data, ax=axs[2])
sns.barplot(x='Christmas', y='Weekly_Sales', data=data, ax=axs[3])
```

```python
Super_Bowl = data[data['Super_Bowl']== 1]['Weekly_Sales'].sum()
Labor_Day = data[data['Labor_Day']== 1]['Weekly_Sales'].sum()
Thanksgiving = data[data['Thanksgiving']== 1]['Weekly_Sales'].sum()
Christmas = data[data['Christmas']== 1]['Weekly_Sales'].sum()

print('Super_Bowl : ',Super_Bowl)
print('Labor_Day : ',Labor_Day)
print('Thanksgiving : ',Thanksgiving)
print('Christmas : ',Christmas)
```

### 1- According to the plots and the numbers calculated the super bowl is the holiday with the most sales and christmas is the one with the least sales

```python
# creating season column
season_month = {
            12:'Winter', 1:'Winter', 2:'Winter',
            3:'Spring', 4:'Spring', 5:'Spring',
            6:'Summer', 7:'Summer', 8:'Summer',
            9:'Autumn', 10:'Autumn', 11:'Autumn'}
data["Season"] = data["Month"].map(season_month)
```

```python
data
```

```python
Winter = data[data['Season']== 'Winter']['Weekly_Sales'].sum()
Spring = data[data['Season']== 'Spring']['Weekly_Sales'].sum()
Autumn = data[data['Season']== 'Autumn']['Weekly_Sales'].sum()
Summer = data[data['Season']== 'Summer']['Weekly_Sales'].sum()

print('Winter : ',Winter)
print('Spring : ',Spring)
print('Autumn : ',Autumn)
print('Summer : ',Summer)
```

```python
sns.barplot(x = ['Winter','Spring','Autumn','Summer'],
            y = [Winter, Spring, Autumn, Summer]
            )
```

### 2-The sales in the summer season is higher than the sales in the winter season (the spring season is the highest overall)

```python
data['Weekly_Sales'].corr(data['Unemployment'])
```

### 3- There's a negative correlation between the weekly sales and unemployement meaning that when the sales increase the unemployement rate decreases

Now i will try to create a regression model to predict the weekly sales

```python
y= data['Weekly_Sales']
```

```python
X=data.drop(['Weekly_Sales', 'Date','Day', 'Month', 'Year'], axis=1)
```

```python
X
```

```python
X = pd.get_dummies(X, columns = ['Season', 'Store'])
X
```

```python
scaler = StandardScaler()
# transform data
X = scaler.fit_transform(X)
print(X)
```

```python
#Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
```

## Regression

### 1- Linear Regression Model

```python
LR = LinearRegression()
LR.fit(X_train, y_train)
```

```python
LR.score(X_test, y_test)
```

### 2- Decision Tree Model

```python
Dtr = DecisionTreeRegressor(random_state = 44)
Dtr.fit(X_train, y_train)
```

```python
Dtr.score(X_test, y_test)
```

### 3- Random Forest Model

```python
Rfr = RandomForestRegressor(n_estimators = 300, random_state = 44)
Rfr.fit(X_train, y_train)
```

```python
Rfr.score(X_test, y_test)
```

## Clustering

```python
Xc = data[['Unemployment','Weekly_Sales']]
```

```python
scaler = StandardScaler()
# transform data
Xc = scaler.fit_transform(Xc)
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=44
)
kmeans.fit(Xc)
```

```python
kmeans.inertia_
```

```python
plt.figure(figsize=(12,12))
ax = sns.scatterplot(x = Xc[:,0], y = Xc[:,1], alpha=0.9, s=400, color='black', hue=kmeans.labels_).plot()
plt.scatter(kmeans.cluster_centers_[ : , 0], kmeans.cluster_centers_[ : , 1], c='red', s=450, alpha=1, marker='o')

# Set x-axis label
plt.xlabel('Unemployment')
# Set y-axis label
plt.ylabel('Weekly_Sales')
plt.show()
```
